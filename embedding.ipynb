```python
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize the Lemmatizer
lem = WordNetLemmatizer()
def clean_text(text):
    text=str(text)
    # Remove content in square brackets (e.g., [Music])
    text = re.sub(r'\[.*?\]', '', text)
    
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Convert text to lowercase
    text = text.lower()
    
    # Tokenize text into words
    words = word_tokenize(text)
    
    # Remove non-alphabetic tokens
    words = [word for word in words if word.isalpha()]
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    # Lemmatization (considering verb form as in your example)
    words = [lem.lemmatize(word, "v") for word in words]
    
    # Join the cleaned tokens back into a string
    cleaned_text = ' '.join(words)
    
    return cleaned_text
```

```python
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Assuming df is already defined with the 'transcript' column
transcripts = list(df['transcript'])

# Define fixed length (in terms of vectorized features)
fixed_feature_count = 4000

# Vectorize function with padding/truncation
def vectorize_and_pad(text, max_features):
    # If the transcript is empty, return a zero vector
    if not text.strip():
        return [0] * max_features  # Return zero vector for empty transcripts
    
    # Vectorize the text using TfidfVectorizer (with max_features to control the size)
    vectorizer = TfidfVectorizer(max_features=max_features)
    tfidf_matrix = vectorizer.fit_transform([text])
    
    # Get the feature vector (it might be shorter than max_features)
    feature_vector = tfidf_matrix.toarray()[0]
    
    # Pad or truncate the feature vector to match max_features
    if len(feature_vector) < max_features:
        # If the feature vector is smaller than max_features, pad it with zeros
        padded_vector = np.pad(feature_vector, (0, max_features - len(feature_vector)), 'constant')
    else:
        # If the feature vector is larger than max_features, truncate it
        padded_vector = feature_vector[:max_features]
    
    return padded_vector

# Apply the transformation to all transcripts in the DataFrame
uniform_features = [
    vectorize_and_pad(t, fixed_feature_count) for t in transcripts
]

# Assign the processed features back to the DataFrame
df['transcript_features'] = uniform_features
```

```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch
from tqdm import tqdm


# Combine the 'title', 'description', and 'tags' into a single column
df['combined_text'] = df['title'] + " " + df['description'] + " " + df['tags']

# Load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Ensure the model is in evaluation mode
model.eval()

# Function to generate embeddings
def generate_embedding(text):
    # Tokenize and encode the text
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=128)
    with torch.no_grad():  # Disable gradient computation for inference
        outputs = model(**inputs)
    # Get the pooled output (embedding)
    embedding = outputs.pooler_output.squeeze().numpy()  # Convert tensor to numpy array
    return embedding

# Apply the embedding generation function to each row
tqdm.pandas()  # For progress bar
df['combined_embedding'] = df['combined_text'].progress_apply(generate_embedding)

# Verify the DataFrame
print(df.head())
```

```python
import numpy as np
from sklearn.decomposition import PCA

# Weights for combined and transcript embeddings
weight_combined = 0.7
weight_transcript = 0.3

# Function to create final_embedding with weightage
def create_weighted_final_embedding(row):
    transcript_emb = row['transcript_features']
    combined_emb = row['combined_embedding']
    
    # Normalize embeddings to avoid scale mismatch
    if np.linalg.norm(transcript_emb) != 0:
        transcript_emb = transcript_emb / np.linalg.norm(transcript_emb)
    
    if np.linalg.norm(combined_emb) != 0:
        combined_emb = combined_emb / np.linalg.norm(combined_emb)
    
    # Reduce the transcript embedding to 768 dimensions using PCA
    pca = PCA(n_components=768)  # Reducing to 768 dimensions
    transcript_emb_reduced = pca.fit_transform(transcript_emb.reshape(1, -1))  # Reshaping for PCA
    transcript_emb_reduced = transcript_emb_reduced.flatten()  # Flatten back to 1D array
    
    # Weighted combination
    final_emb = weight_combined * combined_emb + weight_transcript * transcript_emb_reduced
    return final_emb

# Assuming 'df' contains the relevant embeddings
df['final_embedding'] = df.apply(create_weighted_final_embedding, axis=1)

# Verify the result
print(df.head(5))
```
